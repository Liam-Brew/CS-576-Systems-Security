# Lecture 02: Introduction to the Web

**Uniform Resource Locators (URL)**: dictate where entities are located

Web is organized according to a three-tier architecture: 

1. presentation tier (browsers): retrieve and present data
2. logic tier (servers): serve documents, implement logic, provides plugins for functionality etc.
3. data tier (databases): deal with storing and operating on data

Applications run in all three of these layers, which confuses the security process and facilitates many vulnerabilities

**Hyper Text Transfer Protocol (HTTP)**: stateless protocol to send and receive data. HTTP requests consist of three components:

1. request or response line
2. header section
3. entity body

TCP is used to ensure reliable connections and consists of a three-way handshake:

1. initial request from client
2. acknowledgement of request from server
3. acknowledgement of acknowledgement from client

Passwords are the most commonly used form of web authentication

Session IDs are generated by the server after the first request. It is then sent back to the client, who then includes it in subsequent requests

- Requests cannot be linked based on HTTP alone
